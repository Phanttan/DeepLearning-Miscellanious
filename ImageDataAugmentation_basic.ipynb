# Load Image

```
from google.colab import drive
drive.mount('/content/drive', force_remount=False) 
img_path = '/content/drive/My Drive/Colab Notebooks/datasets/bird.jpg'
img_path
```

```
from keras.preprocessing.image import load_img
img = load_img(img_path)
#img.__dict__
```

```
img._size
```

```
img
```

```
# Convert to Numpy array
from keras.preprocessing.image import img_to_array
data = img_to_array(img)
data.shape
```

```
data[0][0].shape
```

```
# Expand dimension to One sample
from numpy import expand_dims
samples = expand_dims(data, 0)
samples.shape
```

# Create data generator


### With width_shift_range = [-200, 200]

```
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(width_shift_range=[-200,200])
```

```
# Prepare iterator
it = datagen.flow(samples, batch_size=1)
it
```

```
import matplotlib.pyplot as plt

# Generate samples and plots
for idx in range(9):
  # Define subplot
  plt.subplot(330 + 1 + idx)
  # Generate batch of images
  batch = it.next()
  image = batch[0].astype('uint8')
  plt.imshow(image)
plt.show()
```

### With height_shift_range = 0.2


```

datagen = ImageDataGenerator(height_shift_range=0.2)
it = datagen.flow(samples, batch_size=1)
for idx in range(9):
  plt.subplot(330 + 1 + idx)
  batch = it.next()
  image = batch[0].astype('uint8')
  plt.imshow(image)
plt.show()
```

### Random Rotation Augmentation

```
datagen = ImageDataGenerator(rotation_range=90)
it = datagen.flow(samples)
for idx in range(9):
  plt.subplot(330+ 1+ idx)
  batch = it.next()
  image = batch[0].astype('uint8')
  plt.imshow(image)
plt.show()
```

### Shearing

```
datagen = ImageDataGenerator(shear_range=4)
it = datagen.flow(samples)
for idx in range(9):
  plt.subplot(330+ 1+ idx)
  batch = it.next()
  image = batch[0].astype('uint8')
  plt.imshow(image)
plt.show()
```
